```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as web
import numpy as np
import yfinance as yf
import seaborn as sn
from statsmodels.tsa.stattools import adfuller
from sklearn.linear_model import LinearRegression

stocks =yf.download(["AUBANK.NS","AXISBANK.NS","BANDHANBNK.NS","FEDERALBNK.NS","HDFCBANK.NS","ICICIBANK.NS","IDFCFIRSTB.NS","INDUSINDBK.NS","KOTAKBANK.NS","PNB.NS","RBLBANK.NS","SBIN.NS"], 
                    start = '2020-01-01', end = '2021-09-17')
a = stocks['Adj Close']["AUBANK.NS"]
b = stocks['Adj Close']["AXISBANK.NS"]
c = stocks['Adj Close']["BANDHANBNK.NS"]
d = stocks['Adj Close']["FEDERALBNK.NS"]
e = stocks['Adj Close']["HDFCBANK.NS"]
f = stocks['Adj Close']["ICICIBANK.NS"]
g = stocks['Adj Close']["IDFCFIRSTB.NS"]
h = stocks['Adj Close']["INDUSINDBK.NS"]
i = stocks['Adj Close']["KOTAKBANK.NS"]
j = stocks['Adj Close']["PNB.NS"]
k = stocks['Adj Close']["RBLBANK.NS"]
l = stocks['Adj Close']["SBIN.NS"]

data = {'Au-a': a,'Axis-b':b,'Bandhan-c':c,'Federal-d':d,'HDFC-e':e,'ICICI-f':f,'IDFC-g':g,'Indus-h':h,'Kotak-i':i,'PNB-j':j,'RBL-k':k,'SBI-l':l}

df = pd.DataFrame(data,columns=['Au-a','Axis-b','Bandhan-c','Federal-d','HDFC-e','ICICI-f','IDFC-g','Indus-h','Kotak-i','PNB-j','RBL-k','SBI-l'])  
corrMatrix = df.corr()
plt.figure(figsize=(12,8 ))
sn.heatmap(corrMatrix, annot=True,cmap='RdYlGn')
plt.xlabel('X')
plt.ylabel('Y')
plt.show(corrMatrix)
```

```python
import scipy.stats as stats

x =np.array(i).reshape((-1,1))
y= np.array(e)
             
model = LinearRegression().fit(x,y)
y_pred = model.predict(x)
res = y-y_pred

p_value = adfuller(res)
print('p-value:%f'%p_value[1]) 
 
mean = res.mean()
std1 = mean + res.std()
std2 = mean + 2*res.std()
std3 = mean + 3*res.std()
std_1 = mean - res.std()
std_2 = mean - 2*res.std()
std_3 = mean - 3*res.std()
df = pd.DataFrame({'residuals':res,'mean':mean,'std':std1,'std2':std2,'std3':std3,'std_1':std_1,'std_2':std_2,'std_3':std_3})
plt.rcParams['figure.figsize']=(16,6)
df.plot()
plt.title('RESIDUALS')
plt.show()
```
