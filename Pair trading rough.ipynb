```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as web
import numpy as np

stocks = web.get_data_yahoo(['ICICIBANK.NS','HDFCBANK.NS'], start = '2020-01-01', end = '2021-05-20')

icici = stocks['Adj Close']['ICICIBANK.NS'].plot()
hdfc = stocks['Adj Close']['HDFCBANK.NS'].plot()
plt.legend(['icici','hdfc'])
plt.title('closing price')
plt.show()

stocks = web.get_data_yahoo(['ICICIBANK.NS','HDFCBANK.NS'], start = '2020-01-01', end = '2021-05-20')
icici = stocks['Adj Close']['ICICIBANK.NS']
hdfc = stocks['Adj Close']['HDFCBANK.NS']
ratio1 = hdfc/icici
plt.plot(ratio1)
plt.title('ratio hdfc/icici')
plt.show()

icici = stocks['Adj Close']['ICICIBANK.NS']
hdfc = stocks['Adj Close']['HDFCBANK.NS']
corellation = hdfc.corr(icici)
print('ratio corelation =',corellation)

icici = stocks['Adj Close']['ICICIBANK.NS']
hdfc = stocks['Adj Close']['HDFCBANK.NS']
ratio = hdfc/icici
mean = ratio.mean()
std = ratio.std()
print(['std deviation of ratio is =',std],['mean of ratio is =',mean])

icici = stocks['Adj Close']['ICICIBANK.NS']
hdfc = stocks['Adj Close']['HDFCBANK.NS']
ratio = hdfc/icici
ratio.plot.density(color = 'green')
plt.show()
```

```python
stocks = web.get_data_yahoo(['ICICIBANK.NS','HDFCBANK.NS'], start = '2020-01-01', end = '2021-05-20')
icici = stocks['Adj Close']['ICICIBANK.NS']
hdfc = stocks['Adj Close']['HDFCBANK.NS']
ratio1 = hdfc/icici
mean = ratio1.mean()
std1 = mean + ratio1.std()
std2 = mean + 2*ratio1.std()
std3 = mean + 3*ratio1.std()
std_1 = mean - ratio1.std()
std_2 = mean - 2*ratio1.std()
std_3 = mean - 3*ratio1.std()
df = pd.DataFrame({'ratio':ratio1,'mean':mean,'std':std1,'std2':std2,'std3':std3,'std_1':std_1,'std_2':std_2,'std_3':std_3})
plt.plot(df)
plt.title('ratio hdfc/icici')
plt.show()
```
