```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as web
import numpy as np
import yfinance as yf
import seaborn as sn

stocks =yf.download(["SAIL.NS","WELCORP.NS","JINDALSTEL.NS","TATASTEEL.NS","HINDALCO.NS","ADANIENT.NS","NMDC.NS","HINDZINC.NS","JSWSTEEL.NS","VEDL.NS","COALINDIA.NS","MOIL.NS"], 
                    start = "2020-01-01", end = "2021-05-28")
a = stocks['Adj Close']["SAIL.NS"]
b = stocks['Adj Close']["WELCORP.NS"]
c = stocks['Adj Close']["JINDALSTEL.NS"]
d = stocks['Adj Close']["TATASTEEL.NS"]
e = stocks['Adj Close']["HINDALCO.NS"]
f = stocks['Adj Close']["ADANIENT.NS"]
g = stocks['Adj Close']["NMDC.NS"]
h = stocks['Adj Close']["HINDZINC.NS"]
i = stocks['Adj Close']["JSWSTEEL.NS"]
j = stocks['Adj Close']["VEDL.NS"]
k = stocks['Adj Close']["COALINDIA.NS"]
l = stocks['Adj Close']["MOIL.NS"]

data = {'SAIL-a': a,'WELCORP-b':b,'JINDAL-c':c,'TATASTEEL-d':d,'HINDALCO-e':e,'ADANIENT-f':f,'NMDC-g':g,'HINDZINC-h':h,'JSWSTEEL-i':i,'VEDL-j':j,'COALINDIA-k':k,'MOIL-l':l}

df = pd.DataFrame(data,columns=['SAIL-a','WELCORP-b','JINDAL-c','TATASTEEL-d','HINDALCO-e','ADANIENT-f','NMDC-g','HINDZINC-h','JSWSTEEL-i','VEDL-j','COALINDIA-k','MOIL-l'])  
corrMatrix = df.corr()
plt.figure(figsize=(12,8 ))
sn.heatmap(corrMatrix, annot=True,cmap='RdYlGn')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
from sklearn.linear_model import LinearRegression
import scipy.stats as stats

x =np.array(e).reshape((-1,1))
y= np.array(d)
             
model = LinearRegression().fit(x,y)
y_pred = model.predict(x)
res = y-y_pred

p_value = adfuller(res)
print('p-value:%f'%p_value[1]) 
 
mean = res.mean()
std1 = mean + res.std()
std2 = mean + 2*res.std()
std3 = mean + 3*res.std()
std_1 = mean - res.std()
std_2 = mean - 2*res.std()
std_3 = mean - 3*res.std()
df = pd.DataFrame({'residuals':res,'mean':mean,'std':std1,'std2':std2,'std3':std3,'std_1':std_1,'std_2':std_2,'std_3':std_3})
plt.rcParams['figure.figsize']=(16,6)
df.plot()
plt.title('RESIDUALS')
plt.show()
```
